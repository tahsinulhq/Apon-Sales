Problem 1 data preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel('Apon Main Database Gender Data.xlsx')
df2 = pd.read_excel("Month 1.xlsx")
df3 = pd.read_excel("Month 2.xlsx")
df4 = pd.read_excel("Month 3.xlsx")
```

```python
df1.head()
```

```python
df4.head()
```

```python
df1.rename(columns = {"Apon ID": "Customer Code"}, inplace=True)
```

```python
df1.head()
```

```python
df2.rename(columns = {"Total Sales Amount": "Month 1 Sales Amount"}, inplace=True)
df3.rename(columns = {"Total Sales Amount": "Month 2 Sales Amount"}, inplace=True)
df4.rename(columns = {"Total Sales Amount": "Month 3 Sales Amount"}, inplace=True)
```

```python
df2.head()
```

```python
df = pd.merge(df2, df3, on = 'Customer Code', how= 'outer', suffixes= ('_left', '_right'))
```

```python
df
```

```python
df = pd.merge(df, df4, on= 'Customer Code', how= 'outer', suffixes= ('_left', '_right'))
```

```python
df
```

```python
df = pd.merge(df, df1, on = 'Customer Code', how= 'outer', suffixes= ('_left', '_right'))
```

```python
df
```

```python
df.columns
```

```python
df.drop(['SL_left', 'SL_right', 'SL', 'Unnamed: 3','Unnamed: 4','Unnamed: 5','Unnamed: 6','Unnamed: 7'], axis =1, inplace = True)
```

```python
df
```

```python
df['Month 1 Sales Amount'] = df['Month 1 Sales Amount'].fillna(0)
df['Month 2 Sales Amount'] = df['Month 2 Sales Amount'].fillna(0)
df['Month 3 Sales Amount'] = df['Month 3 Sales Amount'].fillna(0)
```

```python
df
```

```python
df['Total Sales Amount'] = df['Month 1 Sales Amount'] + df['Month 2 Sales Amount'] + df['Month 3 Sales Amount']
```

```python
df.head()
```

```python
df = df[df['Total Sales Amount'] !=0.0]
```

```python
df
```

```python
df.insert(6, "Month 1 Customer Status", " ")
df['Month 1 Customer Status'] = np.where(df['Month 1 Sales Amount'] > 0, 'yes', 'no')
df.head()
```

```python
df.insert(7, "Month 2 Customer Status", " ")
df['Month 2 Customer Status'] = np.where(df['Month 2 Sales Amount'] > 0, 'yes', 'no')
df.head()
```

```python
df.insert(8, "Month 3 Customer Status", " ")
df['Month 3 Customer Status'] = np.where(df['Month 3 Sales Amount'] > 0, 'yes', 'no')
df.head()
```

```python
df.to_csv('Apon_Sales_1.csv')
```
